```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
```

```python
food = pd.read_csv("food_ML.csv")
```

```python
y = "Energy_kcal" 
x_1 = food.iloc[0:0, 3:7].columns.values
x_2 = food.iloc[0:0, 7:11].columns.values
x_3 = food.iloc[0:0, 11:15].columns.values
x_4 = food.iloc[0:0, 15:19].columns.values
x_5 = food.iloc[0:0, 19:].columns.values
```

```python
sns.set(style="ticks")
sns.pairplot(food, y_vars=y, x_vars=x_1, hue="FoodGroup");
sns.pairplot(food, y_vars=y, x_vars=x_2, hue="FoodGroup");
sns.pairplot(food, y_vars=y, x_vars=x_3, hue="FoodGroup");
sns.pairplot(food, y_vars=y, x_vars=x_4, hue="FoodGroup");
sns.pairplot(food, y_vars=y, x_vars=x_5, hue="FoodGroup");
```
